```python
# imports
import networkx as nx
import matplotlib.pyplot as plt
import numpy as np

%matplotlib inline
```

```python
# graph values
G = nx.read_pajek('power.paj')
degree_values=np.array(nx.degree(G).values())
N=nx.number_of_nodes(G)

# histogram
bins=np.arange(0.5,20.5)
deg_k=np.histogram(degree_values,bins=bins)

# degree distribution
P_k=deg_k[0]/np.float(N)

# plot
plt.bar(bins[:-1],P_k)
plt.xticks(np.arange(1,18))
plt.xlabel(r'$k$',fontsize=20)
plt.ylabel(r'$P(k)$',fontsize=20)
```

```python
plt.plot(bins[:-1],P_k,'bo')
plt.xscale('log'); plt.yscale('log') # log-log
plt.xlim([1,100]);
```

```python
plt.plot(bins[:-1],1-np.cumsum(P_k),'b-')
plt.xscale('log'); plt.yscale('log');
plt.xlim([1,100]);
```
