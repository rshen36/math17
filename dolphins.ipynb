```python
import networkx as nx
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline
```

```python
G = nx.read_pajek('dolphins.paj')
G = nx.Graph(G)
```

```python
plt.figure(figsize=(12,8))
pos1=nx.spring_layout(G)
nx.draw(G,pos1,node_color=[0.9,0.4,0.8],edge_color='k',width=3,node_size=500,with_labels=True)
```

```python
deg1d = nx.degree_centrality(G)
deg1 = np.array(deg1d.values())
plt.figure(figsize=(12,8))
nx.draw(G,pos1,node_color=[0.9,0.4,0.8],edge_color='k',width=2,
        node_size=15000*deg1,with_labels=True)
```

```python
eig1d = nx.eigenvector_centrality(G)
eig1 = np.array(eig1d.values())
plt.figure(figsize=(12,8))
nx.draw(G,pos1,node_color=[0.9,0.4,0.8],edge_color='k',width=2,
        node_size=15000*eig1,with_labels=True)
```

```python
pgrk1d=nx.pagerank(G)
pgrk1=np.array(pgrk1d.values())
plt.figure(figsize=(12,8))
nx.draw(G,pos1,node_color=[0.9,0.4,0.8],edge_color='k',width=2,
        node_size=15000*pgrk1,with_labels=True)
```

```python
p1_list = list()
p2_list = list()
for k in deg1d:
    if k in eig1d:
        p1_list.append((deg1d[k],eig1d[k]))
    if k in pgrk1d:
        p2_list.append((deg1d[k],pgrk1d[k]))
d1, e1 = zip(*p1_list)
d2, p2 = zip(*p2_list)

fig, (ax1, ax2) = plt.subplots(1,2)
fig.set_figheight(5)
fig.set_figwidth(10)

ax1.scatter(d1,e1,s=200,c='black')
ax1.set_xlim([0, 0.22])
ax1.set_ylim([-0.05, 0.35])
ax1.set_xlabel(r'$C_D^i$ (degree)',fontsize=20)
ax1.set_ylabel(r'$e_i$ (eigenvector)',fontsize=20)
ax2.scatter(d2,p2,s=200,c='black')
ax2.set_xlim([0, 0.22])
ax2.set_ylim([0, 0.04])
ax2.set_xlabel(r'$C_D^i$ (degree)',fontsize=20)
ax2.set_ylabel(r'$p_i$ (PageRank)',fontsize=20)

node_labels = range(0,nx.number_of_nodes(G))
for i, num in enumerate(node_labels):
    ax1.annotate(num, xy=(d1[i],e1[i]),xytext=(0,0),textcoords='offset points',
                 ha='center',va='center',color='white',fontsize=10,family='sans-serif')
    ax2.annotate(num, xy=(d2[i],p2[i]),xytext=(0,0),textcoords='offset points',
                 ha='center',va='center',color='white',fontsize=10,family='sans-serif')
fig.tight_layout()
```

```python
plt.figure(figsize=(12,4))
cls1d=nx.closeness_centrality(G)
cls1=np.array(cls1d.values())
plt.figure(figsize=(12,8))
nx.draw(G,pos1,node_color=[0.9,0.4,0.8],edge_color='k',
        width=2,node_size=15000*cls1,with_labels=True)
```

```python
plt.figure(figsize=(12,4))
bet1d=nx.betweenness_centrality(G)
bet1=np.array(bet1d.values())
plt.figure(figsize=(12,8))
nx.draw(G,pos1,node_color=[0.9,0.4,0.8],edge_color='k',
        width=2,node_size=15000*bet1,with_labels=True)
```

```python
p1_list = list()
p2_list = list()
for k in deg1d:
    if k in cls1d:
        p1_list.append((deg1d[k],cls1d[k]))
    if k in bet1d:
        p2_list.append((deg1d[k],bet1d[k]))
d1, c1 = zip(*p1_list)
d2, b2 = zip(*p2_list)

fig, (ax1, ax2) = plt.subplots(1,2)
fig.set_figheight(5)
fig.set_figwidth(10)

ax1.scatter(d1,c1,s=200,c='black')
ax1.set_xlim([0, 0.22])
ax1.set_ylim([.15, 0.45])
ax1.set_xlabel(r'$d_i$ (degree)',fontsize=20)
ax1.set_ylabel(r'$C_i$ (closeness)',fontsize=20)
ax2.scatter(d2,b2,s=200,c='black')
ax2.set_xlim([0, 0.22])
ax2.set_ylim([-0.05, 0.3])
ax2.set_xlabel(r'$d_i$ (degree)',fontsize=20)
ax2.set_ylabel(r'$b_i$ (betweenness)',fontsize=20)

node_labels = range(0,nx.number_of_nodes(G))
for i, num in enumerate(node_labels):
    ax1.annotate(num, xy=(d1[i],c1[i]),xytext=(0,0),textcoords='offset points',
                 ha='center',va='center',color='white',fontsize=10,family='sans-serif')
    ax2.annotate(num, xy=(d2[i],b2[i]),xytext=(0,0),textcoords='offset points',
                 ha='center',va='center',color='white',fontsize=10,family='sans-serif')
fig.tight_layout()
```
